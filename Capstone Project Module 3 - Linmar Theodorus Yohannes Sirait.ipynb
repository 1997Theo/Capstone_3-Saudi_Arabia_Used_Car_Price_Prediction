## **Business Problem Understanding**

**Context**

Syarah merupakan sebuah perusahaan platform marketplace jual-beli mobil bekas yang berlokasi di Riyadh, Saudi Arabia. Dalam proses jual-beli pada syarah, interaksi terjadi diantara penjual mobil dan pembeli secara langsung, sedangkan Syarah hanya berfungsi sebagai wadah untuk mempertemukan, dan juga memberikan jaminan keamanan sehingga pembeli dapat bertransaksi tanpa takut terkena penipuan. Hal ini dikarenakan setiap penjual yang ingin menjual mobilnya lewat platform Syarah, maka tim Syarah akan melakukan inspeksi secara langsung terhadap mobil bekas dan juga melakukan background check terhadap calon penjual. 

Untuk menjual mobil pada platform Syarah, calon penjual cukup mendaftarkan akun, kemudian mengisi data diri dan juga mobil yang ingin dijual, serta memasukkan beberapa dokumentasi mobil yang akan dijual. Setelah itu, tim Syarah akan melakukan inspeksi terhadap mobil yang akan dijual dan juka syarat-syarat administrasi dan hasil inspeksi telah memenuhi syarat, penjual bisa langsung menentukan harga mobil yang akan dijual dan memilih paket iklan pada platform Syarah. Pada platform Syarah ini, penjual dibebaskan untuk memberikan harga pada mobil yang akan dijual, dimana pada beberapa kasus penjual kesulitan dalam menentukan harga mobil tersebut. Jika penjual memberikan harga yang jauh lebih mahal dibandingkan harga pasaran, maka tentu saja mobil tersebut akan sulit terjual dan resource (uang yang dikeluarkan penjual untuk iklan dan waktu yang dihabiskan) akan terbuang begitu saja, sebaliknya jika penjual memasang harga jual yang terlampau murah, maka tentu saja penjual tidak akan mendapatkan profit sebanyak yang seharusnya didapatkan.

Melansir dari [Mordor Intelligence](https://www.mordorintelligence.com/industry-reports/saudi-arabia-used-car-market/market-size#:~:text=Saudi%20Arabia%20Used%20Car%20Market%20was%20valued%20at%20USD%204.91,7.36%25%20over%20the%20forecast%20period.), pada tahun 2021, market mobil bekas Saudi Arabia divaluasi 4.91 miliar USD, dan diprediksi akan menembus 8.69 miliar USD pada akhir tahun 2027, memastikan CAGR sebesar 7.36% secara solid sampai 2027. Oleh karena itu bisnis model seperti Syarah ini akan menjadi bisnis model yang sangat menggiurkan, hal itu menyebabkan tingginya persaingan antara platform mobil bekas yang ada di saudi arabia. Untuk dapat bertahan pada bisnis ini, setiap platform harus memiliki cara tertentu agar penjual dan pembeli nyaman untuk bertransaksi pada platform tersebut, salah satunya adalah dengan memberikan rekomendasi harga jual kepada penjual, sehingga harga yang ada platform tersebut menjadi relevan, dan pembeli juga secara otomatis akan memilih platform tersebut untuk bertransaksi

**Problem Statement**

Karena harga jual mobil ditentukan langsung oleh penjual, terkadang penjual mengalami kebingungan dalam menentukan harga yang pantas untuk mobil bekas yang akan dijual, dimana terdapat kemungkinan bahwa penjual akan memberikan harga yang sangat tinggi sehingga harga-harga mobil yang ada pada platform akan menjadi kurang relevan terhadap pasar, sehingga pembeli juga menjadi enggan untuk bertransaksi di platform dan pada akhirnya platform tersebut akan menjadi sepi dan ada kemungkinan untuk bangkrut.

Cara paling jitu untuk menjadikan platform Syarah menjadi platform nomor satu adalah `Sistem rekomendasi harga mobil jual yang akurat`, agar penjual mendapatkan profit yang sepadan dan pembeli juga mendapatkan mobil yang sesuai selera dengan harga yang 'reasonable' (masuk akal). 
Hal ini sangat dibutuhkan agar baik penjual dan pembeli mempercayai perusahaan sebagai `perusahaan yang capable dalam menciptakan kondisi market yang tidak overprice maupun underprice sehingga baik penjual dan pembeli tidak akan dirugikan.`


**Goals**

Agar dapat mewujudkan sistem rekomendasi yang relevan, perusahaan membutuhkan suatu tools untuk dapat memprediksi harga mobil bekas seakurat mungkin dengan harga pasar berdasarkan data-data pada mobil tersebut seperti Tahun keluaran mobil, Ukuran Mesin mobil, Fitur pada mobil, Tipe mobil, Perusahaan Pembuat mobil, Jarak Tempuh mobil, Asal mobil, Region mobil yang dijual, dan Transmisi pada mobil. Adanya perbedaan pada fitur-fitur tersebut sudah pasti menyebabkan perbedaan pada harga jual mobil.

Syarah berharap dengan adanya sistem rekomendasi yang memprediksi harga secara fair, maka jumlah user yang aktif pada platform akan bertambah. Sehingga tentu saja hal itu akan menaikkan revenue perusahaan, dalam konteks ini 'paid ads' yang dipurchase oleh pembeli

**Analytic Approach**

Hal yang perlu dilakukan untuk menciptakan sistem rekomendasi yang akurat adalah dengan menganalisis fitur-fitur yang terdapat pada data mobil bekasnya, dan menganalisis seberapa berpengaruhnya suatu fitur terhadap pembentukan harga mobil bekas.

Untuk dapat melakukannya, kita akan membuat suatu model regresi yang akan menjadi tools prediksi harga yang akan direkomendasikan ke calon penjual mobil nantinya agar penjual mobil dapat memberikan harga yang menghasilkan keuntungan yang 'reasonable' sehingga pembeli juga tertarik untuk membeli dan platform, dan Gross Transaction Value (GTV) perusahaan akan semakin besar.

**Metric Evaluation**

Untuk melakukan evaluasi pada model yang dibuat,  metrik yang akan digunakan adalah RMSE, MAE, dan MAPE, di mana RMSE adalah nilai rata-rata akar kuadrat dari error, MAE adalah absolut nilai rata-rata dari error, sedangkan MAPE adalah rata-rata persentase error yang dihasilkan oleh model regresi. Semakin kecil nilai RMSE, MAE, dan MAPE yang dihasilkan, berarti model semakin akurat dalam memprediksi harga sewa sesuai dengan limitasi fitur yang digunakan. Hanya saja pada kasus ini, kita akan `menekankan penilaian pada metrik MAE dan MAPE`, dimana dari sisi business owner ataupun penjual mobil, `akan lebih mudah menginterpretasikan error dalam bentuk persen ataupun nilai toleransinya` terhadap harga jual aktual.

Selain itu, kita juga bisa menggunakan nilai R-squared atau adj. R-squared jika model yang nanti terpilih sebagai final model adalah model linear. Nilai R-squared digunakan untuk mengetahui seberapa baik model dapat merepresentasikan varians keseluruhan data. Semakin mendekati 1, maka semakin fit pula modelnya terhadap data observasi. Namun, metrik ini tidak valid untuk model non-linear.

**Visualisasi Kolom Numerikal**

```python
# Plot distribusi data numerikal pada dataset, beserta uji hipotesa dengan menggunakan normaltest
plt.figure(figsize=(20,15))
for kolom in numerical_features:

    fig, (ax1, ax2) = plt.subplots(2, sharex = True, figsize = (7.5,5), 
                                    gridspec_kw={"height_ratios": (.15, .85),
                                                "hspace": 0.05})

    stat, pvalue = normaltest(df[kolom].dropna())

    if pvalue > 0.05 :
        distribusi = f'pvalue = {pvalue:.5f}, data terdistribusi normal'
    else :
        distribusi = f'pvalue = {pvalue:.5f}, data TIDAK terdistribusi normal'

    ax1.set_title(f"Distribution of {kolom}\n{distribusi}", color = "black")
    sns.boxplot(x = df[kolom], ax = ax1)
    ax1.set(xlabel = "")

    sns.histplot(data = df[kolom], ax = ax2, stat = "probability", kde=True, bins=50)

    plt.show()
```

### **Handling Outliers**

```python
# Membuat list kolom numerikal yang akan menjadi feature
numerical_features = df_model.describe().columns
```

```python
# mencari nilai lower dan upper fence pada masing-masing kolom numerikal

list_lower = []
list_upper = []
number_lower = []
number_inside = []
number_upper = []
percentage_outlier = []

for feature in numerical_features: 
    IQR = df_model[feature].quantile(0.75) - df_model[feature].quantile(0.25)
    lower_fence = df_model[feature].quantile(0.25) - 1.5*IQR
    upper_fence = df_model[feature].quantile(0.75) + 1.5*IQR
    list_lower.append(lower_fence)
    list_upper.append(upper_fence)
    number_lower.append(len(df_model[df_model[feature]<lower_fence]))
    number_inside.append(len(df_model[(df_model[feature]>=lower_fence) & (df_model[feature]<=upper_fence)]))
    number_upper.append(len(df_model[df_model[feature]>upper_fence]))
    percentage_outlier.append(round((len(df_model[df_model[feature]<lower_fence]) + len(df_model[df_model[feature]>upper_fence])) / len(df_model)  *100, 2))

outliers = pd.DataFrame({
    'Kolom' : numerical_features,
    'Lower' : list_lower,
    'Upper' : list_upper,
    '# Lower' : number_lower,
    '# Inside' : number_inside,
    '# Upper' : number_upper,
    '% Outlier' : percentage_outlier
})

outliers
```

Dari sini kita dapat melihat bahwa untuk setiap kolom numerikal yang akan dijadikan feature, terdapat data outlier. Untuk memutuskan Outlier mana yang lebih pantas dihapus, maka setiap kolom akan dikaji secara terpisah dan dianalisis

### **Encoding**

Agar dapat menggunakan fitur kategorikal pada machine learning, fitur-fitur tersebut terlebih dahulu harus diubah menjadi data numerikal, sebelumnya mari kita lihat nilai-nilai pada kolom kategorikal :

```python
# menampilkan jumlah unik dan unique values untuk setiap kolom
pd.set_option('display.max_colwidth', None)
# data unik di tiap kolom
listItem = []
for col in df_model.describe(include='object').columns :
    listItem.append( [col, df_model[col].nunique(), df_model.sort_values(col)[col].unique()])

tabel1Desc = pd.DataFrame(columns=['Column Name', 'Number of Unique', 'Unique Values'],
                     data=listItem)
tabel1Desc
```

dari enam fitur kategorikal diatas, akan lakukan proses encoding sesuai dengan karakteristiknya masing-masing, berikut penjabaran setiap kolom dan encoder yang akan digunakan :
1. `Type` : Karena nilai pada kolom Type tidak memiliki urutan dan jumlah nilai uniknya ada banyak (>5), maka diputuskan untuk memakai `binary encoder`
2. `Region` : Karena nilai pada kolom Region tidak memiliki urutan dan jumlah nilai uniknya ada banyak (>5), maka diputuskan untuk memakai `binary encoder`
3. `Make`: Karena nilai pada kolom Make tidak memiliki urutan dan jumlah nilai uniknya ada banyak (>5), maka diputuskan untuk memakai `binary encoder`
4. `Gear_Type` : Pada kolom ini terdapat 2 nilai unik yaitu automatic dan manual, dimana untuk jenis mobil yang sama, mobil dengan transmisi automatic sudah pasti memiliki harga yang lebih mahal dibandingkan mobil dengan transmisi manual karena kemudahan berkendara, oleh karena itu diputuskan untuk memakai `ordinal encoder` karena memiliki urutan
5. `Origin` : Karena nilai pada kolom Origin tidak memiliki urutan dan jumlah nilai uniknya sedikit (<5), diputuskan untuk memakai `onehot encoder`
6. `Options` : Pada kolom ini terdapat 3 nilai yaitu Full, Semi Full, dan Standard, dimana untuk mobil dengan tipe yg sama, mobil dengan fitur full pasti lebih mahal dibandingkan yang Semi Full, dan Semi Full lebih mahal dari mobil dengna fitur Standard. Oleh karena itu diputuskan untuk memakai `ordinal encoder` karena memiliki urutan

Kita akan menggunakan Column untuk melakukan proses encoding sebagai berikut :

```python
# ordinal mapping
ordinal_mapping = [
    {'col':'Gear_Type', 'mapping':{'Manual' : 1, 'Automatic' : 2, None:0}},
    {'col':'Options', 'mapping':{'Standard' : 1, 'Semi Full' : 2, 'Full' : 3, None:0}},
] 
```

```python
# Mengubah fitur kategorikal menjadi numerikal

transformer = ColumnTransformer([
    ('One Hot', OneHotEncoder(drop='first'), ['Origin']),
    ('ordinal', OrdinalEncoder(cols=['Gear_Type','Options'], mapping=ordinal_mapping), ['Gear_Type','Options']),
    ('Binary', BinaryEncoder(), ['Type','Region','Make'])
], remainder='passthrough')
```

### **Choose the Best Benchmark Model**

Untuk menentukan model terbaik mana yang akan dipakai, kita akan melakukan cross validation dengan berbagai macam model listrik, kemudian nantinya akan memilih 3 model dengan scoring terbaik untuk di predict ke test set untuk melihat apakah hasil masih konsisten atau ada perubahan. Oleh karena itu diputuskan untuk menggunakan skala logaritmik yang nantinya akan diinverse agar dapat diinterpretasi.

```python
# Define algoritma yang digunakan
linreg = LinearRegression()
ridge = Ridge(random_state=0)
lasso = Lasso(random_state=0)
knn = KNeighborsRegressor()
dtree = DecisionTreeRegressor(random_state=0)
rf = RandomForestRegressor(random_state=0)
adaboost = AdaBoostRegressor(random_state=0)
gradboost = GradientBoostingRegressor(random_state=0)
xgboost = XGBRegressor(random_state=0)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_linreg = TransformedTargetRegressor(linreg, func=np.log, inverse_func=np.exp)
log_ridge = TransformedTargetRegressor(ridge, func=np.log, inverse_func=np.exp)
log_lasso = TransformedTargetRegressor(lasso, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dtree = TransformedTargetRegressor(dtree, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_adaboost = TransformedTargetRegressor(adaboost, func=np.log, inverse_func=np.exp)
log_gradboost = TransformedTargetRegressor(gradboost, func=np.log, inverse_func=np.exp)
log_xgboost = TransformedTargetRegressor(xgboost, func=np.log, inverse_func=np.exp)
```

`Untuk model linear regression, kita akan menambahkan intercept pada X_train saat di cross validation` agar memberikan model yang lebih akurat, jika nantinya linear regression memiliki score paling tinggi, kita juga akan memasukkan intercept pada X_testnya. Selain itu untuk model `linear regression, KNN, Ridge, dan Lasso akan digunakan robust scaler` agar model memberikan hasil yang lebih akurat, sedangkan `model dengan tree-based tidak menggunakan scaler karena tidak diperlukan.`

```python
X_train2 = sm.add_constant(X_train)
```

```python
# Scaling data
scaler = RobustScaler()

# Kandidat algoritma
models = [log_linreg, log_ridge, log_lasso, log_knn, log_dtree, log_rf, log_adaboost, log_gradboost, log_xgboost]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

# Mencari algoritma terbaik berdasarkan tiap metrics
for i in models:
    
    kfold = KFold(n_splits=5)

    # Pipeline dengan scaler untuk model linreg, ridge, lasso, KNN
    estimator1 = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', i)
    ])

    # Pipeline tanpa scaler untuk model tree-based
    estimator2 = Pipeline([
        ('preprocessing', transformer),
        ('model', i)
    ])

    if i in [linreg,ridge,lasso,knn] :
        estimator = estimator1
    else :
        estimator = estimator2

    if i == linreg :
        X_used = X_train
    else :
        X_used = X_train2

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator, 
        X_used, 
        y_train, 
        cv=kfold, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
        )

    print(model_cv_rmse, i)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_used, 
        y_train, 
        cv=kfold, 
        scoring='neg_mean_absolute_error', 
        error_score='raise'
        )

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_used, 
        y_train, 
        cv=kfold, 
        scoring='neg_mean_absolute_percentage_error', 
        error_score='raise'
        )
    
    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())
```

Sebelum memutuskan model terbaik untuk dituning, ada baiknya dilakukan validasi sekali lagi terhadap ketiga model tersebut dengan menggunakan test set. Apabila tidak ada perubahan dalam perankingan, akan digunakan model XGBoost sebagai benchmark model terbaik yang kemudian akan dilakukan Hyperparameter Tuning untuk mendapatkan hasil terbaik dari model tersebut

```python
# Benchmark 2 model terbaik
models = {
    'XGBoost': TransformedTargetRegressor(xgboost, func=np.log, inverse_func=np.exp),
    'RandomForest': TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp),
    'GradientBoost' : TransformedTargetRegressor(gradboost, func=np.log, inverse_func=np.exp)
}

score_rmse = []
score_mae = []
score_mape = []

# Prediksi pada test set
for i in models:
    # tidak perlu pakai scaler karena based modelnya adalah decision tree
    model = Pipeline([
        ('preprocessing', transformer),
        ('model', models[i])
        ])

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score_rmse.append(np.sqrt(mean_squared_error(y_test, y_pred)))
    score_mae.append(mean_absolute_error(y_test, y_pred))
    score_mape.append(mean_absolute_percentage_error(y_test, y_pred))

score_before_tuning = pd.DataFrame({'RMSE': score_rmse, 'MAE': score_mae, 'MAPE': score_mape}, index=models.keys())
score_before_tuning
```

**Analisis**

Setelah dilakukan predict terhadap test set, didapati bahwa `XGBoost Regressor` dengan MAPE 17.9% masih menjadi model terbaik, oleh karena itu diputuskan bahwa model XGBoost Regressor adalah model benchmark terbaik, yang kemudian akan dilakukan Hyperparameter Tuning untuk mendapatkan hasil yang maksimal.

**Penjelasan Mengenai Model XGBoost Regressor**

XGBoost merupakan model yang dikembangkan dari model Decision Tree yang merupakan metode pembelajaran yang melibatkan grafik seperti pohon untuk memodelkan data yang bersifat kontinu atau kategorik. Decision Tree juga terdiri dari serangkaian pertanyaan biner yang ketika dijawab akan menghasilkan prediksi tentang data. 

XGBoost (Extreme Gradient Boosting) merupakan salah satu metode boosting dan merupakan pemutakhiran dari metode Gradient Boosting, dimana kumpulan Decision Tree yang pembangunan pohon berikutnya akan bergantung pada pohon sebelumnya (secara sekuensial). Prediksi akan dimulai dengan nilai rata-rata pada dataset yang kemudian akan dilihat nilai residualnya, dan kemudian akan dibuat model decision tree untuk memprediksi nilai residual tersebut. Untuk menghindari overfitting terhadap data train, tree tersebut akan dikalikan dengan suatu skala yang disebut learning rate. Setelah pohon pertama ditambahkan kedalam prediksi, maka nilai residualnya akan semakin berkurang, dan untuk mengantisipasinya, akan dibuat pohon kedua untuk mengurangi nilai residual tersebut, dan begitu seterusnya sampai didapatkan model yang akurat.

Perbedaan utama antara GradientBoosting dengan XGBoost adalah XGbost menggunakan teknik regularisasi di dalamnya. Dengan kata sederhana, ini adalah bentuk reguler dari algoritme peningkatan gradien yang ada, sehingga XGBoost tidak mudah untuk overfitting dan tetap memberikan hasil yang maksimal pada 'test set'-nya. Selain itu, peningkatan yang dilakukan pada XGBoost dibandingkan GradientBoost adalah XGBoost mempertimbangkan potensi kerugian untuk semua kemungkinan pemisahan untuk membuat cabang baru. XGBoost mengatasi ketidakefisienan ini dengan melihat distribusi fitur di semua titik data dalam sebuah 'leaf' pada tree dan menggunakan informasi ini untuk mengurangi ruang pencarian kemungkinan pemisahan fitur. ### **Hyperparameter Tuning**

Karena parameter pada model XGBoost Regressor ada banyak, diputuskan untuk menggunakan randomized search dengan iterasi sebanyak 1000 kali. Selain itu, proses tuning akan dilakukan sebanyak dua kali dimana untuk tuning kedua, parameternya akan dipersempit sesuai dengan 5 best parameter pada tuning pertama, dengan harapan dapat menghasilkan hasil yang maksimal.

 **Hyperparameter Tuning tahap 1**

```python
# Kedalaman pohon
max_depth = list(np.arange(1, 11))

# Learning rate
learning_rate = list(np.arange(1, 100)/100)

# Jumlah pohon
n_estimators = list(np.arange(100, 201))

# Jumlah baris tiap pohon (% dari total baris train set)
subsample = list(np.arange(2, 10)/10)

# Gamma (min_impurity_decrease)
gamma = list(np.arange(1, 11)) # Semakin besar nilainya, semakin konservatif/simpel modelnya

# Jumlah feature yang digunakan untuk tiap pohon (% dari total kolom train set)
colsample_bytree = list(np.arange(1, 10)/10)

# Alpha (regularization)
reg_alpha = list(np.logspace(-3, 3, 7)) # Semakin besar nilainya, semakin konservatif/simpel modelnya


# Hyperparam space XGboost
hyperparam_space_xgb = {
    'model__regressor__max_depth': max_depth, 
    'model__regressor__learning_rate': learning_rate,
    'model__regressor__n_estimators': n_estimators,
    'model__regressor__subsample': subsample,
    'model__regressor__gamma': gamma,
    'model__regressor__colsample_bytree': colsample_bytree,
    'model__regressor__reg_alpha': reg_alpha
}
```

```python
# Benchmark model dengan hyperparameter tuning
xgb = XGBRegressor(random_state=0)
log_xgb = TransformedTargetRegressor(regressor=xgb, func=np.log, inverse_func=np.exp)

# Membuat algorithm chains
estimator_log_xgb = Pipeline([
        ('preprocessing', transformer),
        ('model', log_xgb),
        ])

crossval = KFold(n_splits=5)

# Hyperparameter tuning
random_xgb = RandomizedSearchCV(
    estimator_log_xgb, 
    param_distributions = hyperparam_space_xgb,
    n_iter = 1000,
    cv = crossval, 
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'], 
    n_jobs = -1,
    refit = 'neg_mean_absolute_error', # Hanya bisa memilih salah stau metric untuk optimisasi
    random_state = 0  
)
```

```python
# Fitting data training untuk mencari parameter terbaik
random_xgb.fit(X_train, y_train)
```

```python
# Menyimpan 5 kombinasi parameter yang menghasilkan score terbaik, yang akan digunakan untuk hyperparameter tahap 2 nantinya
dftop5params=pd.DataFrame(random_xgb.cv_results_).sort_values(by=['rank_test_neg_root_mean_squared_error', 'rank_test_neg_mean_absolute_error', 'rank_test_neg_mean_absolute_percentage_error']).head()
```

```python
# Cek skor dan params terbaik
print('XGBoost')
print('Best_score:', random_xgb.best_score_)
print('Best_params:', random_xgb.best_params_)
```

    - Jika memungkinkan, sebaiknya ditambahkan fitur yang berisi metode pembayaran yang diterima oleh penjual (apakah cash atau cicilan). Hal ini sudah pasti mempengaruhi harga mobil yang akan dijual karena ada pengaruh dari nilai uang itu sendiri (faktor inflasi dll).

- **Jika memungkinkan, sebaiknya pada data ditambahkan kolom user (penjual) agar kita dapat melihat penjual mana yang telah menjual mobil paling banyak, sehingga kita dapat melakukan strategic partnership dengan user tersebut untuk dapat mengejar target perusahaan, yaitu memiliki Gross Transaction Value setinggi mungkin.**

- **Agar dapat memberikan rekomendasi yang akan membantu penjual merumuskan harga yang lebih akurat, dapat dilakukan analisis karakteristik pada data yang mempunyai prediksi overestimate dan data yang mempunyai prediksi underestimate.
